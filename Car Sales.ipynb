## Cleaning Data

##### renaming Column

```python
df.rename(columns={'Price_in_thousands':'price'},inplace=True)
df.rename(columns={'Sales_in_thousands':'Sales'},inplace=True)
df.rename(columns={'__year_resale_value':'year resale value'},inplace=True)
```

#### check  if there are Missing value

```python
df.isnull().sum()
```

### Fill Missing Value

```python
def fill_missing_Value(df):
    for column in df.columns:
        if df[column].isnull().any():
            mean_value = df[column].mean()
            df[column].fillna(mean_value, inplace=True)
    return df
fill_missing_Value(df)
```

##### Check if there are duplicate values

```python
df.duplicated().sum()
```

# Analyzing Data

## 1- How many Model are made by each manufacturer

```python
Car_Man =df.groupby('Manufacturer')['Model'].nunique().to_frame()
Car_Man = Car_Man.sort_values(by='Model' , ascending=False)
Car_Man 
```

## 2- The best cars have an Power_perf_factor

```python
Data = df.groupby('Manufacturer')['Power_perf_factor' ].max().to_frame()
Data = Data.sort_values(by='Power_perf_factor' , ascending=False)
Data
```

## 3- Top 15 Model by Sales

```python
data = df.sort_values(by = 'Sales' , ascending=False)
data = data[['Model' ,'Sales']].head(15)
data
```

## 4 - TOP 10 Cars by Sales

```python
ss= df.groupby('Manufacturer')['Sales'].sum().reset_index()
ss=ss.sort_values(by = 'Sales' , ascending=False)[:10]
ss
```

## 5- Average price and sales per manufacturer

```python
df.groupby('Manufacturer')[['Sales' , 'price']].mean()
```
